```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
irisclsf = tree.DecisionTreeClassifier()
irisclsf = irisclsf.fit(iris.data, iris.target)
```

```python
import graphviz
```

```python
dot_data = tree.export_graphviz(
    irisclsf,
    out_file=None,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled=True,
    rounded=True,
    special_characters=True
)
graph = graphviz.Source(dot_data)
graph.render("iris")
```
