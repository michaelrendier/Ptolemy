```python
from lxml import html as HTML
import requests
```

```python
def main():
    r = requests.get("https://httpbin.org")
    html_source = r.text
    root = HTML.fromstring(html_source)
    
    page_title = root.xpath('/html/head/title/text()')[0]
    print("PAGE TITLE", page_title)
    
if __name__ == "__main__":
    main()
```

```python
import string

def divide_list(l, n):
    for i in range(0, len(l), n):
        yield l[i: i + n]
        
your_name = list(input("Input Name? ").upper())

for i in your_name:
    if your_name.count(i) > 1:
        while your_name.count(i) > 1:
            your_name.remove(i)

if "I" in your_name:
    your_name[your_name.index("I")] = "IJ"

if "J" in your_name:
    your_name[your_name.index("J")] = "IJ"

for i in string.ascii_uppercase:
    if i in your_name:
        pass
    elif i in your_name and i == "I" or i == "J":
        i = 'IJ'
    elif i == "I":
        your_name.append("IJ")
    elif i == "J":
        pass
    else:
        your_name.append(i)
        
if your_name.count("IJ") > 1:
    your_name.pop(your_name[your_name.index("IJ") + 1:].index("IJ") + your_name.index("IJ") + 1)        
        
square = list(divide_list(your_name, 5))
for i in square: print(i)
```
