```python
import multiprocessing as mp
import numpy as np
from time import time
from timeit import timeit
```

```python
def howmmany_within_range(row, minimum, maximum):
    count = 0
    for n in row:
        if minimum <= n <= maximum:
            count += 1
    return count
```

```python
# Prepare Data
np.random.RandomState(100)
arr = np.random.randint(0, 10, size=[100000, 5])
data = arr.tolist()
data[:5]
```

```python
# Solution without parallelization
results = []
def myfunc():
    for row in data:
        results.append(howmmany_within_range(row, minimum=4, maximum=8))
```
