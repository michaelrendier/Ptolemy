```python
import os
from subprocess import Popen, call
import polyglot
from polyglot.text import Text, Word

mystring = Text("West of House You are standing in an open field west of a white house, with a boarded front door. There is a small mailbox here.")
```

```python
print(mystring.pos_tags)
```

```python
keywords = [[word, tag] for word, tag in mystring.pos_tags if tag in ['PROPN', 'NOUN']]
```

```python
print(keywords)
```
