```python
class House:
    
    def __init__(self, *args, **kwargs):
        super().__init__()
        self.value = args[0]
        self.parent = kwargs['parent']
        self.address = "1234 some street"
        self.house_sq_ft = 45 * 62 #45 feet by 62 feet
        
    def other_func(self):
        print('House.other_func() called')
        
class Rental(House):
    
    def __init__(self, *args, **kwargs):
        super().other_func()
        self.args = [arg for arg in args]
        self.objdict = kwargs
        print("Inside Rental:", self.args)
        print("Inside Rental:", self.objdict)
        
thisone = Rental('window title', parent=None)

print("Rental.__dict__:", thisone.__dict__)
```
