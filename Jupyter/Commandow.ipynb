```python
def rows(data, n):
    """Yield successive n-sized chunks from data"""
    for i in range(0, len(data), n):
        yield data[i:i + n]
```

```python
def stamp(data):
    returned = []
    try:
        data[0] + 0
        
        for i in range(len(data)):
            returned.append(data[i])
        returned.append("\n")
    
    except TypeError:
        
        for i in range(len(data)):
            for j in range(len(data[i])):
                returned.append(data[i][j])
            returned.append("\n")
            
    return "".join(returned)
        
        
```
